# **Ploting bar chart to show the population trend of India over years**


*Importing* the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

*reading the csv file*

```python
df=pd.read_csv("world_population.csv")
df
```

*filtering the data of india from the dataset*

```python
ind_data=df[df['Country/Territory']=='India']
```

```python
years=['2022 Population',	'2020 Population',	'2015 Population','2010 Population',	'2000 Population',
       '1990 Population',	'1980 Population',	'1970 Population']
```

```python
ind_population=ind_data[years].values.flatten()
```

```python
int_years=[int(year.split()[0]) for year in years]
```

*plotting the graph*

```python
plt.figure(figsize=(15, 8))
plt.bar(int_years, ind_population,width=1.5,color='red')
plt.title('Population of India Over the Years')
plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Population in billion')
plt.xticks(int_years)
plt.show()
```
